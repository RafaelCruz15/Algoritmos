# Permutaciones
## Trabajando con una muestra aleatoria

Definimos el algoritmo de Bubble Sort, aplicando los contadores de instrucciones, comparaciones e intercambios, e importando las librerias necesarias

```python
import itertools
import random
import matplotlib.pyplot as pl

# algoritmo de bubble sort
def bubblesort(A):
    intercambios = 0  # intercambioes
    comparaciones = 0  # comparaciones
    instrucciones = 0  # instrucciones del algoritmo
    for passnum in range(len(A) - 1, 0, -1):
        instrucciones += 1
        for i in range(passnum):
            instrucciones += 1
            if A[i] > A[i + 1]:
                comparaciones += 1
                instrucciones += 1
                temp = A[i]
                A[i] = A[i + 1]
                A[i + 1] = temp
                intercambios += 3
    return (A, intercambios, comparaciones, instrucciones)
```

Definimos la funcion promedio

```python
def promedio(A):
    return ((sum(A))/len(A))
```

Luego llenamos un arreglo con numeros del 1 al 9, para trabajar con un n = 9 en las permutaciones y luego las invocamos

```python
arreglo = []
# trabajamos con un tamaÃ±o de n = 9
for i in range(1, 10):
    arreglo.append(i)

# cantidad de permutaciones 9! = 362.880
p = list(itertools.permutations(arreglo, 9))
```

Creamos y capturamos en arreglos las variables intercambios, comparaciones e instrucciones, con una muestra de 200000 permutaciones

```python
intercambios_array = []
comparaciones_array = []
instrucciones_array = []

# elegimos trabajar con 200000 permutaciones
N = 200000

for i in range(N):

    # posicion aleatoria de las permutaciones
    j = random.randint(0,len(p)-1)

    # permutacion
    p_all = p[j]
    #print(p_all)

    # creamos un arreglo para la tupla aleatoria
    n_arr = []
    for i in range(len(p_all)):
        n_arr.append(p_all[i])

    solucion = (bubblesort(n_arr))

    intercambios_array.append(solucion[1])
    comparaciones_array.append(solucion[2])
    instrucciones_array.append(solucion[3])
```

Calculamos los promedios correspondientes

```python
print("Promedios")
print("El promedio de intercambios es: " , promedio(intercambios_array))
print("El promedio de comparaciones es: " , promedio(comparaciones_array))
print("El promedio de instrucciones es: " , promedio(instrucciones_array))
```

Luego creamos las distribuciones de cada variable

```python
# distribucion de intercambios
distr_inter = []
cont_dis_inter = []
a = min(intercambios_array)
b = max(intercambios_array)
for i in range(a, b+1):
    distr_inter.append(i)
    cont_dis_inter.append(intercambios_array.count(i))
    i += 1

# distribucion de comparaciones
distr_compar = []
cont_dis_compar = []
a = min(comparaciones_array)
b = max(comparaciones_array)
for i in range(a, b+1):
    distr_compar.append(i)
    cont_dis_compar.append(comparaciones_array.count(i))
    i += 1

# distribucion de instrucciones
distr_instru = []
cont_dis_instru = []
a = min(instrucciones_array)
b = max(instrucciones_array)
for i in range(a, b+1):
    distr_instru.append(i)
    cont_dis_instru.append(instrucciones_array.count(i))
    i += 1
```

Y mostramos las distribuciones de cada variable

```python
print("Distribucion de intercambios")
for i in range(len(cont_dis_inter)):
    print(distr_inter[i], " -->", cont_dis_inter[i])
print("\n")

print("Distribucion de comparaciones")
for i in range(len(cont_dis_compar)):
    print(distr_compar[i], " -->", cont_dis_compar[i])
print("\n")

print("Distribucion de instrucciones")
for i in range(len(cont_dis_instru)):
    print(distr_instru[i], " -->", cont_dis_instru[i])
print("\n")
```

Graficando cada una

```python
pl.plot(distr_inter,cont_dis_inter)
pl.xlabel('Numero de Intercambios');
pl.ylabel('Cantidad que se repiten');
```

```python
pl.plot(distr_compar,cont_dis_compar)
pl.xlabel('Numero de Comparaciones');
pl.ylabel('Cantidad que se repiten');
```

```python
pl.plot(distr_instru,cont_dis_instru)
pl.xlabel('Numero de Instrucciones');
pl.ylabel('Cantidad que se repiten');
```
