# Fibonacci en Python

```python
import matplotlib.pyplot as pl
import numpy as np
import time

print("Serie Fibonacci para 100000 datos")

def fibonacci(n):
    resultado = 0
    b = 1
    for i in range(n):
        c = resultado + b
        b = resultado
        resultado = c
    #print("El resultado de " , n , " es de " , resultado)

# tama√±o de los datos de mil en mil
n = []
for i in range(0,101000,1000):
    n.append(i)
#print(n)

# calculo de fibonacci para 1000 hasta 101000
t = []
for i in range(0,101000,1000):
    start_time = time.time()
    fibonacci(i)
    tiempo = (time.time() - start_time)
    t.append(tiempo)
    #print("El tiempo para " , i , " es de " , tiempo)

pl.plot(t,n)
pl.xlabel('Tiempo (segundos)');
pl.ylabel('Tamano de los Datos (n)');
```
