# Matriz en Python

```python
import numpy as np
import matplotlib.pyplot as pl
import random, time

# crea la matriz cuadrada de un tamaño y valor cualquiera
def array(tam, valor):
    A = []
    i = 0
    j = 0

    # llena la matriz de 0's
    for i in range(0, tam):
        A.append([0] * tam)

    # llena la matriz de el valor que digiten
    for i in range(0, tam):
        for j in range(0, tam):
            A[i][j] = valor

    return A

n = []
# arreglo de tamaño n = 50, 100, 150, 200, ... , 500
for i in range(50, 550, 50):
    n.append(i)

# almacena los valores para cada n
t_50 = []
t_100 = []
t_150 = []
t_200 = []
t_250 = []
t_300 = []
t_350 = []
t_400 = []
t_450 = []
t_500 = []

# 15 ejecuciones para cada n
for i in range(15):
    for j in range(50, 550, 50):
        # matriz de tamaño jxj, con un valor de 1's y de 2's
        matriz1 = array(j, 1)
        matriz2 = array(j, 2)
        # con numpy
        matriz1 = np.array(matriz1)
        matriz2 = np.array(matriz2)

        start_time = time.time()  # tiempo inicial

        # multiplicacion
        np.dot(matriz1, matriz2)

        tiempo = (time.time() - start_time)

        # pregunta donde esta j para almacenar el tiempo de acuerdo a n
        if(j == 50):
            t_50.append(tiempo)
        elif (j == 100):
            t_100.append(tiempo)
        elif (j == 150):
            t_150.append(tiempo)
        elif (j == 200):
            t_200.append(tiempo)
        elif (j == 250):
            t_250.append(tiempo)
        elif (j == 300):
            t_300.append(tiempo)
        elif (j == 350):
            t_350.append(tiempo)
        elif (j == 400):
            t_400.append(tiempo)
        elif (j == 450):
            t_450.append(tiempo)
        elif (j == 500):
            t_500.append(tiempo)

        #print("El tiempo para ", j, " es de ", tiempo)

# sacamos el promedio y almacenamos en un nuevo arreglo
t_final = []
t_final.append(np.mean(t_50))
t_final.append(np.mean(t_100))
t_final.append(np.mean(t_150))
t_final.append(np.mean(t_200))
t_final.append(np.mean(t_250))
t_final.append(np.mean(t_300))
t_final.append(np.mean(t_350))
t_final.append(np.mean(t_400))
t_final.append(np.mean(t_450))
t_final.append(np.mean(t_500))
#print(t_final)

# grafica de los tiempos promedios
pl.plot(t_final,n)
pl.xlabel('Tiempo (segundos)');
pl.ylabel('Tamano de los Datos (n)');

print("Tiempos promedios con servidor web")
for i in range(10):
    print("Para n = " , n[i] , " tenemos un tiempo promedio de ",  t_final[i])
    
```
