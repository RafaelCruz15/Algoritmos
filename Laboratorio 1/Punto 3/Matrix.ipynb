# Matriz en Python

```python
import numpy as np
import matplotlib.pyplot as pl
import random, time

# crea la matriz cuadrada
def array(tam, valor):
    A = []
    i = 0
    j = 0

    # llena la matriz de 0's
    for i in range(0, tam):
        A.append([0] * tam)

    # llena la matriz de el valor que digiten
    for i in range(0, tam):
        for j in range(0, tam):
            A[i][j] = valor

    return A

# total de datos a recibir
tamano = 400

print("Multiplicacion de matrices cuadradas de tamano 1 x 1 hasta " , tamano , " x " , tamano)

n = []
# arreglo de tamaño
for i in range(tamano):
    n.append(i+1)
#print(n)

t = []
for i in range(1,tamano+1):
    # matriz de tamaño 2x2, con un valor de 1's y de 2's
    matriz1 = array(i,1)
    matriz2 = array(i,2)
    # con numpy
    matriz1 = np.array(matriz1)
    matriz2 = np.array(matriz2)

    start_time = time.time()  # tiempo inicial

    # multiplicacion
    np.dot(matriz1, matriz2)

    tiempo = (time.time() - start_time)
    t.append(tiempo)
    #print("El tiempo para ", i, " es de ", tiempo)

#print(t)

# grafica
pl.plot(t,n)
pl.xlabel('Tiempo (segundos)');
pl.ylabel('Tamano de los Datos (n)');

print("tiempo de ejecucion para 10 valores usando un servidor web")
for i in range(0,tamano + (tamano/10),tamano/10):
    # matriz de tamaño 2x2, con un valor de 1's y de 2's
    matriz1 = array(i,1)
    matriz2 = array(i,2)
    # con numpy
    matriz1 = np.array(matriz1)
    matriz2 = np.array(matriz2)

    start_time = time.time()  # tiempo inicial

    # multiplicacion
    np.dot(matriz1, matriz2)

    tiempo = (time.time() - start_time)
    print("El tiempo para ", i, " es de ", tiempo)
    
```
