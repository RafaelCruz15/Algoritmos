# Muestreo

Vamos a probar algunos datos aleatorios usando una función de numpy.

```python
# muestra 100 puntos aleatorios con una media de 0 y un estándar de 1. Esta es una distribución normal estándar.
X = np.random.normal(0, 1, 100)
print(X)
```

# Agregar etiquetas de ejes

Ningún quant que se respete deja un gráfico sin hachas etiquetadas. Aquí hay algunos comandos para ayudar con eso.

```python
X = np.random.normal(0, 1, 100)
X2 = np.random.normal(0, 1, 100)

plt.plot(X);
plt.plot(X2);
# eje x
plt.xlabel('Tiempo') # Los datos que generamos no tienen unidades, pero no olvidar las unidades en general.
# eje y
plt.ylabel('Resultados')
plt.legend(['X', 'X2']);
```

# Generando estadísticas

Usemos numpy para tomar algunas estadísticas simples.

```python
# Media aritmética
np.mean(X)
```

```python
# Desviación estándar
np.std(X)
```

# Obteniendo Datos de Precios Reales

Los datos de muestra aleatoria pueden ser geniales para probar ideas, pero obtengamos algunos datos reales. Podemos usar get_pricing para hacer eso. Puedes usar el signo de ? sintaxis como se discutió anteriormente para obtener más información sobre los argumentos de get_pricing.

```python
# Precios reales
data = get_pricing('MSFT', start_date='2012-1-1', end_date='2015-6-1')
print(data)
```

Nuestros datos ahora son un marco de datos. Puede ver el índice de fecha y hora y las columnas con diferentes datos de precios.

```python
data
```

Este es un marco de datos de pandas, por lo que podemos indexar para obtener un precio como este. Para obtener más información sobre pandas, haga [click aqui](http://pandas.pydata.org/pandas-docs/stable/10min.html).

```python
# filtramos la columna "price"
X = data['price']
print(X)
```

Debido a que ahora también hay información de fecha en nuestros datos, proporcionamos dos series a .plot. X.index nos da el índice de fecha y hora y X.values nos da los valores de fijación de precios. Estos se usan como las coordenadas X e Y para hacer un gráfico.

```python
# grafica
plt.plot(X.index, X.values)
plt.ylabel('Price')
plt.legend(['MSFT']);
```

Podemos obtener estadísticas nuevamente sobre datos reales.

```python
# media aritmetica
np.mean(X)
```

```python
# desviación estandar
np.std(X)
```

# Obtener devoluciones de los precios

Podemos usar la función pct_change para obtener devoluciones. Observe cómo soltamos el primer elemento después de hacer esto, ya que será NaN (nada -> algo resulta en un cambio de porcentaje de NaN).

```python
# devolviendo precios
R = X.pct_change()[1:]
print(R)
```

Podemos trazar la distribución de devoluciones como un histograma.

```python
# histograma
plt.hist(R, bins=20)
plt.xlabel('Resultados')
plt.ylabel('Frecuencia')
plt.legend(['Resultados MSFT']);
```

Obtenga estadísticas de nuevo.

```python
# media aritmetica
np.mean(R)
```

```python
# desviación estandar
np.std(R)
```

Ahora retrocedamos y generemos datos de una distribución normal utilizando las estadísticas que estimamos a partir de las declaraciones de Microsoft. Veremos que tenemos buenas razones para sospechar que las devoluciones de Microsoft pueden no ser normales, ya que la distribución normal resultante parece muy diferente.

```python
# Grafica distribucion normal
plt.hist(np.random.normal(np.mean(R), np.std(R), 10000), bins=20)
plt.xlabel('Resultados')
plt.ylabel('Frecuencia')
plt.legend(['Resultados de la distribucion normal']);
```

# Generando una media móvil

La libreria pandas tiene algunas buenas herramientas que nos permiten generar estadísticas progresivas. Aquí hay un ejemplo. Observe cómo no hay un promedio móvil durante los primeros 60 días, ya que no tenemos 60 días de datos para generar la estadística.

```python
# Generando una media movil con pandas
MAVG = pd.rolling_mean(X, window=60)
plt.plot(X.index, X.values)
plt.plot(MAVG.index, MAVG.values)
plt.ylabel('Price')
plt.legend(['MSFT', '60-day MAVG']);
```
